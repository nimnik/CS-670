```python
import pandas as pd

dataf = pd.read_csv("Combined_News_DJIA.csv")

print(len(dataf))
```

```python
s = []
for i in range(1, 26):
    s += ["Top" + str(i)]
print(s)

data_list = []

for i in range(25):
    data_list.append(dataf[s[i]].tolist())

for i in range(25):
    for j in range(len(dataf)):
        if type(data_list[i][j]) == float:
            data_list[i][j] = "None"
        else:
            data_list[i][j] = data_list[i][j][2:-1]
```

```python
number_of_features = 25

topn = []

for j in range(len(dataf)):
  x = ""

  for i in range(number_of_features - 1):
    x += data_list[i][j] + " "
  x += data_list[number_of_features - 1][j]
  topn.append(x)

print(len(topn))
```

```python
key_dict = {}
value_dict = {}

all_words_str = topn[0]

for i in range(1, len(topn)):
  all_words_str += " " + topn[i]

all_words = all_words_str.split()

print(len(all_words))

for word in all_words:
  if word not in key_dict:
    key_dict[word] = 0

for word in all_words:
  key_dict[word] += 1

for word in key_dict:
  if key_dict[word] not in value_dict:
    value_dict[key_dict[word]] = [word]
  else:
    value_dict[key_dict[word]].append(word)

print(len(value_dict))
```

```python
word_count_list = list(value_dict.keys())
word_count_list.sort(reverse=True)

top15_words = []
top15_words_count = []

for i in range(15):
  top15_words.append(value_dict[word_count_list[i]][0])
  top15_words_count.append(word_count_list[i])

print(top15_words)
print(top15_words_count)
```

```python
import matplotlib.pyplot as plt

plt.bar(top15_words, top15_words_count)
plt.xlabel("Words")
plt.ylabel("Count")
plt.title("Top 15 Words")
plt.show()
```

```python
word_count = []
word_count_dict = {}

for i in range(25):
  for j in range(len(dataf)):
    y = len(data_list[i][j].split())
    if y not in word_count:
      word_count.append(y)

for i in range(max(word_count)):
  word_count_dict[i+1] = 0

for i in range(25):
  for j in range(len(dataf)):
    z = len(data_list[i][j].split())
    word_count_dict[z] += 1


word_count_x = []
word_count_y = []

for i in range(1, max(word_count) + 1):
  word_count_x.append(i)
  word_count_y.append(word_count_dict[i])

plt.xlabel("Word Count")
plt.ylabel("Number of News")
plt.title("Number of News per Word Count")

plt.plot(word_count_x, word_count_y)
plt.show()
```

```python
char_count = []

for word in all_words:
  if len(word) not in char_count:
    char_count.append(len(word))

print(char_count)

char_count_dict = {}

for i in range(1, max(char_count) + 1):
  char_count_dict[i] = 0

for word in all_words:
  char_count_dict[len(word)] += 1

char_count_x = []
char_count_y = []

for i in range(1, max(char_count) + 1):
  char_count_x.append(i)
  char_count_y.append(char_count_dict[i])


plt.xlabel("Character Count")
plt.ylabel("Number of Words")
plt.title("Number of Words per Character Count")

plt.plot(char_count_x, char_count_y)
plt.show()
```

```python
target_count = 0

for i in range(len(dataf)):
  if dataf["Label"][i] == 1:
    target_count += 1

target_x = ["0", "1"]
target_y = [len(dataf) - target_count, target_count]

plt.bar(target_x, target_y)
plt.xlabel("Label")
plt.ylabel("Number of Days")
plt.show()
```
